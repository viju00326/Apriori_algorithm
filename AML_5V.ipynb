```python
import numpy as np
import random

# Reward Matrix
R = np.array([
    [-1, -1, -1, -1, 0, -1],
    [-1, -1, -1, 0, -1, 100],
    [-1, -1, -1, 0, -1, -1],
    [-1, 0, 0, -1, 0, -1],
    [0, -1, -1, 0, -1, 100],
    [-1, 0, -1, -1, 0, 100]
])

# Initialize Q-table with zeros
Q = np.zeros((6, 6))

gamma = 0.8
episodes = 1000

# Training the Q-learning model
for i in range(episodes):

    # Random starting state
    state = random.randint(0, 5)

    while state != 5:

        # Possible actions
        possible_actions = []

        for j in range(6):
            if R[state, j] >= 0:
                possible_actions.append(j)

        # Random action selection
        action = random.choice(possible_actions)

        # Next state
        next_state = action

        # Update Q-value
        Q[state, action] = R[state, action] + gamma * np.max(Q[next_state])

        # Move to next state
        state = next_state

# Display Q-table
print("Q Table:")
print(Q)

# Find optimal path
state = 2
print("\nOptimal path:")

while state != 5:
    print(state, end=" -> ")
    next_action = np.argmax(Q[state])
    state = next_action

print("Goal")
```
