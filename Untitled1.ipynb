```python
# Install required library
# pip install gymnasium numpy
!pip install -U gymnasium

import gymnasium as gym
import numpy as np
import random

# Create CartPole environment
env = gym.make('CartPole-v0')

# Q-table size
state_bins = [10, 10, 10, 10]
action_size = env.action_space.n

# Initialize Q-table
q_table = np.zeros(state_bins + [action_size])

# Hyperparameters
learning_rate = 0.1
discount_factor = 0.95
epsilon = 1.0
epsilon_decay = 0.995
min_epsilon = 0.01
episodes = 500

# Function to convert continuous state to discrete
def discretize_state(state):
    upper_bounds = [4.8, 5, 0.418, 5]
    lower_bounds = [-4.8, -5, -0.418, -5]

    ratios = [(state[i] - lower_bounds[i]) /
              (upper_bounds[i] - lower_bounds[i])
              for i in range(len(state))]

    new_state = [int(round((state_bins[i] - 1) * ratios[i]))
                 for i in range(len(state))]

    new_state = [min(state_bins[i] - 1,
                 max(0, new_state[i]))
                 for i in range(len(state))]

    return tuple(new_state)

# Training loop
for episode in range(episodes):
    observation, info = env.reset()
    state = discretize_state(observation)

    done = False
    total_reward = 0

    while not done:
        # Choose action
        if random.uniform(0, 1) < epsilon:
            action = env.action_space.sample()
        else:
            action = np.argmax(q_table[state])

        # Take action
        next_observation, reward, terminated, truncated, _ = env.step(action)
        done = terminated or truncated
        next_state = discretize_state(next_observation)

        # Update Q-table
        best_next_action = np.argmax(q_table[next_state])

        q_table[state][action] += learning_rate * (
            reward +
            discount_factor *
            q_table[next_state][best_next_action] -
            q_table[state][action]
        )

        state = next_state
        total_reward += reward

    # Reduce exploration
    epsilon = max(min_epsilon, epsilon * epsilon_decay)

    if episode % 50 == 0:
        print(f"Episode {episode}, Reward: {total_reward}")

print("Training Finished!")

# Testing the trained model
observation, info = env.reset()
done = False

while not done:
    env.render()

    state = discretize_state(observation)
    action = np.argmax(q_table[state])

    next_observation, reward, terminated, truncated, _ = env.step(action)
    done = terminated or truncated
    observation = next_observation

env.close()
```
